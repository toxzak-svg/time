**Task:** `temporal_ambiguity`  |  **Metric:** Accuracy

Dataset: `/kaggle/input/datasets/zacharymaronek/adversarial-temporal-crusher-dataset/temporal_ambiguity.csv`

## Task Definition

Queries asking "what is currently true about X?" where the most recent
fact differs from the currently valid fact.

```python
# @title Cell 2: Load + Normalize
import pandas as pd

df = pd.read_csv("/kaggle/input/datasets/zacharymaronek/adversarial-temporal-crusher-dataset/temporal_ambiguity.csv")
for col in df.columns:
    df[col] = df[col].apply(lambda x: str(x) if pd.notna(x) else "")
print(f"{len(df)} questions")
print(df.head(2).to_string())
```

```python
# @title Cell 3: Define Task
import kaggle_benchmarks as kbench

@kbench.task(name="temporal_ambiguity")
def temporal_ambiguity(llm, question_id: str, prompt: str, correct_answer: str, naive_answer: str) -> bool:
    response = llm.prompt(f"{prompt}\n\nAnswer with only the value. Do not explain.")
    return correct_answer.lower() in response.lower()
```

```python
# @title Cell 4: Evaluate
results = temporal_ambiguity.evaluate(llm=[kbench.llm], evaluation_data=df)
rdf = results.as_dataframe()
print(f"Accuracy: {rdf['result'].mean()*100:.1f}%")
```

```python
# @title Cell 5: Save
rdf.to_csv("/content/submission_temporal_ambiguity.csv", index=False)
print(f"Saved {len(rdf)} rows")
```
