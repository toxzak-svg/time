# Task 1: Temporal Reversion

**Task:** `temporal_reversion`  |  **Metric:** Accuracy

Dataset: `/kaggle/input/datasets/zacharymaronek/temporal_reversion.csv`

## Task Definition

A fact timeline reverts to a previous state (A→B→A). Query asks about
the **current** period. Correct answer is NOT the most recent mention by timestamp.

## kbench

`@kbench.task(name="temporal_reversion")` → `bool` (aggregates to accuracy)

```python
# @title Cell 2: Load Dataset
import pandas as pd
df = pd.read_csv("/kaggle/input/datasets/zacharymaronek/temporal_reversion.csv")
print(f"{len(df)} questions")
```

```python
# @title Cell 3: Define Task
import kaggle_benchmarks as kbench

@kbench.task(name="temporal_reversion")
def temporal_reversion(llm, question_id: str, prompt: str, correct_answer: str, naive_answer: str) -> bool:
    response = llm.prompt(f"{prompt}\n\nAnswer with only the value. Do not explain.")
    return str(correct_answer).lower() in response.lower()
```

```python
# @title Cell 4: Evaluate
results = temporal_reversion.evaluate(llm=[kbench.llm], evaluation_data=df)
rdf = results.as_dataframe()
print(f"Accuracy: {rdf['result'].mean()*100:.1f}%")
```

```python
# @title Cell 5: Save
rdf.to_csv("/content/submission_temporal_reversion.csv", index=False)
print(f"Saved {len(rdf)} rows")
```
